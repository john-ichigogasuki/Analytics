```python
import MeCab, os
import pandas as pd

# tagger = mecab.Tagger("-Owakati")
tagger = MeCab.Tagger("-Owakati -r /opt/homebrew/etc/mecabrc")
```

```python
input_alpha_filename = 'alpha_Koga.csv'
input_beta_filename = 'beta_Koga.csv'

alpha_path = os.path.join('./TalkData', input_alpha_filename)
beta_path = os.path.join('./TalkData', input_beta_filename)

alpha_data = pd.read_csv(alpha_path)
beta_data = pd.read_csv(beta_path)
```

```python
def calculate_unique_words(text):
    words = tagger.parse(text).strip().split()
    return len(set(words))
```

```python
alpha_data['UniqueWords'] = alpha_data['Detail'].apply(calculate_unique_words)
beta_data['UniqueWords'] = beta_data['Detail'].apply(calculate_unique_words)
```

```python
alpha_avg = alpha_data["UniqueWords"].mean()
beta_avg = beta_data["UniqueWords"].mean()
```

```python
average_unique_alpha_words = alpha_data.groupby('Speaker')['UniqueWords'].mean()
print(average_unique_alpha_words)
average_unique_beta_words = beta_data.groupby('Speaker')['UniqueWords'].mean()
print(average_unique_beta_words)
```

## Word len分析

```python
import pandas as pd
import glob

def addDetailLenColumn(path):
    csv_files = glob.glob(path)
    if not csv_files:
        print(f"指定されたパスに一致するファイルが見つかりません: {path}")
        return
    
    for file in csv_files:
        print(f"処理中のファイル: {file}")
        try:
            df = pd.read_csv(file)
        except Exception as e:
            print(f"ファイルの読み込みに失敗しました: {file}, エラー: {e}")
            continue
        
        if "Detail" not in df.columns:
            print(f"'Detail'列が存在しません: {file}")
            continue
        
        # 'Detail'列の各行の文字数を計算して新しい列 'DetailLen' として追加
        df["WordsLen"] = df["Detail"].apply(len)
        
        try:
            df.to_csv(file, index=False, encoding='utf-8-sig')
            print(f"上書き保存しました: {file}")
        except Exception as e:
            print(f"ファイルの保存に失敗しました: {file}, エラー: {e}")

path = "/Users/john/Programs/University/Study/Analytics/Analyzed/AnalyzedData/analyzed_*_*.csv"
addDetailLenColumn(path)
```

#### Column移動

```python
import pandas as pd
import glob

def move_words_len_column(path):
    # 指定されたパスに一致するすべてのCSVファイルを取得
    csv_files = glob.glob(path)
    
    for file_path in csv_files:
        print(f"処理中のファイル: {file_path}")
        
        # CSVファイルを読み込む
        df = pd.read_csv(file_path)
        
        # 'WordsLen'列を2番目の位置に移動
        cols = df.columns.tolist()
        if 'WordsLen' in cols:
            cols.insert(2, cols.pop(cols.index('WordsLen')))
            df = df[cols]
        
            # 結果を保存
            df.to_csv(file_path, index=False, encoding='utf-8-sig')
            print(f"上書き保存しました: {file_path}")
        else:
            print(f"'WordsLen'列が存在しません: {file_path}")

# 指定されたパスに一致するすべてのCSVファイルに対して操作を行う
path = "/Users/john/Programs/University/Study/Analytics/Analyzed/AnalyzedData/analyzed_*_*.csv"
move_words_len_column(path)
```
